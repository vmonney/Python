Let's import the datasets and have a quick overview:

````python
import pandas as pd
import datetime
import functools
%matplotlib inline
import matplotlib.pyplot as plt

# Import the inflation data
inflation = pd.read_csv('Data/cpi.csv', parse_dates=['DATE'])
# fix auto-parsing date
def fix_date(x):
    if x.year > 2021:
        year = x.year - 100
    else:
        year = x.year
    return datetime.date(year,x.day, x.month)
inflation['DATE'] = inflation['DATE'].apply(fix_date)
inflation.set_index('DATE', inplace=True)
inflation.index = pd.to_datetime(inflation.index)

# Import PPI data
ppi = pd.read_csv('Data/ppi.csv', index_col=0)
ppi.rename(columns={"PIEATI01CHM661N_PC1": "ppi"}, inplace=True)
ppi.index = pd.to_datetime(ppi.index)

# Import Oil Price
oil = pd.read_csv('Data/oil_price.csv',index_col=0)
oil.rename(columns={"DCOILBRENTEU_PC1": "oil"}, inplace=True)
oil.index = pd.to_datetime(oil.index)

# Import expected inflation
expinf = pd.read_csv('Data/expinf12.csv', index_col=0, parse_dates = ['DATE'], dayfirst=True)
expinf.rename(columns={"EXPINF12": "exp_inf"}, inplace=True)
expinf.index = pd.to_datetime(expinf.index)

# Import Real exchange rate
rer = pd.read_csv('Data/RER.csv', index_col=0)
rer.rename(columns={"RBCHBIS_PC1": "exchange_rate"}, inplace=True)
rer.index = pd.to_datetime(rer.index)

# Import output gap
output_gap = pd.read_csv('Data/total_prod.csv', index_col=0)
output_gap.rename(columns={"CHEPROINDQISMEI": "output_gap"}, inplace=True)
output_gap.index = pd.to_datetime(output_gap.index)

# Import unemployment rate
un_rate = pd.read_csv('Data/u_rate.csv', index_col=0, parse_dates=['DATE'])
un_rate.rename(columns={"LMUNRRTTCHM156S": "un_rate"}, inplace=True)
un_rate.index = pd.to_datetime(un_rate.index)

# Merge all data together
data = pd.concat([inflation, ppi, oil, expinf, rer, output_gap, un_rate], axis=1)
data
````

### (c) Please tell us your detailed machine learning strategy 

#### Autoregression (AR)
The autoregression method models the next step in the sequence as a linear function of the observations at prior time steps. We set a parameter __p__ (number of auto-regressive terms) which incorporates past values i.e. lags of dependant variable.

For that we will use the [AR function](https://www.statsmodels.org/v0.12.2/generated/statsmodels.tsa.ar_model.AR.html) from the statsmodels library.

#### ARIMA model

in the Arima model there are 3 parameters that are used to help model the major aspects of a times series: seasonality, trend and noise, that can be decomposed using the [seasonal_compose()](https://www.statsmodels.org/dev/generated/statsmodels.tsa.seasonal.seasonal_decompose.html) function. These parameters are labeled p,d, and q.

- Number of AR (Auto-Regressive) terms (p): lags of dependent variable.
- Number of Differences (d): the amount of differencing to apply to a time series.
- Number of MA (Moving Average) terms (q): q is size of the moving average part of window of the model. We can use the [auto_arima](http://alkaline-ml.com/pmdarima/) function to find the best parameters for us.

#### Tensorflow RNN with LSTM layer

Recurrent neural network (RNN) are a class of neural networks that is powerful for modeling sequence data such as time series. In this particular case, we will add the LSTM layer which stand for Long Short-Term Memory. This layer help us build recurrent models without having to make difficult configuration choices.

Tensorflow LSTM is a layer that we can add in our recurrent neural network (RNN) with Keras to better predict time Serie data. LSTM stand for Long Short-Term Memory. It is a layer that has feedback connections. It is composed of a cell, an input gate, an output gate and a forget gate. The cell remembers values over arbitrary time intervals and the three gates regulate the flow of information into and out of the cell. LSTM networks are well-suited to making predictions based on time series data, since there can be lags of unknown duration between important events in a time series.


#### Preprocessing steps
I will firstly clean my data and rescale them in order to have the same amount of magnitude for each feature. Then I will split the data into train, validation and test set (70%-15%-15%). Be careful here, Since we talk about timeseries, I can't split it randomly. I will take the latest data as validation and test set and train the data with oldest values.

#### Baseline model
I wil build the baseline model for the project with a naive first-order univariate autoregressive process (AR(1)). This model take the following form:
> $\pi_{t+1} = \alpha + \beta \pi_t + \epsilon_{t+1}$

Where $\pi_t$ represent inflation at time period $t$, $\alpha$ is a constant, $\beta < 1$ is the persistence and $\epsilon_{t+1}$ is the estimation error.

#### Metrics
A number of statistical measures will be used to evaluate the models. The statistical measures used in this project are Root Mean Square Error __(RMSE)__, Mean Absolute Error __(MAE)__ and Co-efficient of Determination __(R2)__.

__RMSE__ is a measure of the square root of the mean square error. It is a standard metric for measuring the error in machine learning models.

__R2__ shows how well the independent variables can predict the variance in the dependent variable, in this case the inflation. It is an indicator of how well the model fits the data.

__MAE__ indicates an average value of difference between the predicted values and the actual values.
