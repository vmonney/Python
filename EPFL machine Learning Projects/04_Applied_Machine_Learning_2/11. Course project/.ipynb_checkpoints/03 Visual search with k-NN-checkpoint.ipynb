# Visual search with k-NN
In this section, we will use the KNN algorithm to create a search engine based on the high-level image representation. 
So we will put an image as input, and the algorithm will return the 10 most similar images (according to its parameters). This could be comparable to the google image search engine for example.

## Fit and tune a KNN classifier
First of all, let's import the data and set a baseline model.

```python
import os
import pandas as pd
import numpy as np

# Import train data
with np.load(os.path.join('processed_data', 'features_tr.npz'), allow_pickle=False) as data:
    features = dict(data.items())
    
# Create X/y arrays
X_tr = features['x']
y_tr = features['y']

# import validation data
with np.load(os.path.join('processed_data', 'features_va.npz'), allow_pickle=False) as data:
    features_va = dict(data.items())
    
# Create X/y arrays
X_va = features_va['x']
y_va = features_va['y']


#import raw test Images
with np.load(os.path.join('processed_data', 'features_te.npz'), allow_pickle=False) as data:
    features_te = dict(data.items())
    
# Create X/y arrays
X_te = features_te['x']
y_te = features_te['y']

# Proportion of each image
pd.value_counts(y_tr, normalize=True)
```

```python
from sklearn.dummy import DummyClassifier

# Create the dummy classifier
dummy = DummyClassifier(strategy='most_frequent')

# Fit it to the train data
dummy.fit(X_tr, y_tr)

# Compute accuracy on the test set
accuracy = dummy.score(X_va, y_va)
print('Baseline accuracy: {:.3f}'.format(accuracy))
```

Our baseline accuracy is around 24%. Can the model be improved? Let's try to create a better one.

```python
from sklearn.neighbors import KNeighborsClassifier

# Create k-NN classifier
knn = KNeighborsClassifier(
    n_neighbors=5,
    # Use the simple 'brute' strategy to find nearest neighbors.
    # It's faster in this case!
    algorithm='brute'
)

knn.fit(X_tr, y_tr)

# Generate a set of k values
k_values = np.arange(1, 20, 1)

# Run several times the evaluation
gs_results = []
# Grid search
for k in k_values:
    # Fit k-NN model
    knn.set_params(n_neighbors=k)
    knn.fit(X_tr, y_tr)

    # Save model and its performance on train/test sets
    gs_results.append({
        'k': k,
        'train_accuracy': knn.score(X_tr, y_tr),
        'test_accuracy': knn.score(X_va, y_va)
    })

# Convert results to DataFrame
gs_results = pd.DataFrame(gs_results)
gs_results
```

Much better! We are around 92% of accuracy with the validation set. Let's try to visualize the different results, according to the number k (number of neighbors taken into account.)

```python
%matplotlib inline
import matplotlib.pyplot as plt

# Group results by alpha value
grouped = gs_results.groupby('k')

# Compute train/test mean scores with std
mean_tr = grouped.train_accuracy.mean()
mean_va = grouped.test_accuracy.mean()
std_tr = grouped.train_accuracy.std()
std_va = grouped.test_accuracy.std()

# Plot mean scores
plt.plot(k_values, mean_tr, label='train')
plt.plot(k_values, mean_va, label='validation')

# Add marker for best score
best_k = mean_va.idxmax()
plt.scatter(best_k, mean_va.max(), marker='x', c='red', zorder=10)

# Quantify variance with Â±std curves
plt.fill_between(k_values, mean_tr-std_tr, mean_tr+std_tr, alpha=0.2)
plt.fill_between(k_values, mean_va-std_va, mean_va+std_va, alpha=0.2)
plt.title('Best k: {} with {:.1f}% accuracy'.format(best_k, 100*mean_va[best_k]))
plt.ylabel('accuracy')
plt.xlabel('k')
plt.legend()
plt.show()
```

Our best result seems to be around 5-6 neighbors with an accuracy of 92.8%.

## Pick an image from the test set and plot its 10 nearest neighbors from the train set

Now that we have our model, we can create our search engine. The principle is simple, we select an image at random, and our model returns the 10 closest points (images) which resemble the most to our image.
Let's see what it gives.

```python
# Select a number from 1 to 50 (image number of the test set)
image_test = 49
```

```python
from sklearn.neighbors import NearestNeighbors

# Fit the 10 nearest neighbors for each image in the train set
neigh_10 = NearestNeighbors(n_neighbors=10)
neigh_10.fit(X_tr)

# Extract the distance and neighbors id for the test image
distance, neighbors_id = neigh_10.kneighbors([X_te[image_test]])

# Create a function to flatten the list of lists
def flatten(t):
    return [item for sublist in t for item in sublist]

# flatten the list of lists to have only one list
distance = flatten(distance)
neighbors_id = flatten(neighbors_id)
```

```python
#import raw test Images
with np.load(os.path.join('processed_data', 'test_data.npz'), allow_pickle=False) as data:
    test_image = dict(data.items())
    
#import raw test Images
with np.load(os.path.join('processed_data', 'train_data.npz'), allow_pickle=False) as data:
    train_image = dict(data.items())


# Plot the test image and their 10 neares neighbors
plt.imshow(test_image['x'][image_test])
plt.title("Test image")
plt.axis('off');

plt.figure(figsize = (20,2))
for i, idx in enumerate(neighbors_id):
    plt.subplot(1, 10, i+1)
    plt.imshow(train_image['x'][idx])
    plt.title(round(distance[i], 1))
    plt.axis('off')
```

Our search engine is quite good to find similar vehicles, also when they are in different point of view. You can try another test by changing the image number.
