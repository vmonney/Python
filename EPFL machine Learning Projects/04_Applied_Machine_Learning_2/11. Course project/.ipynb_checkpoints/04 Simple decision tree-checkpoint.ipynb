# Simple decision trees

This time, we will see if we can do better than the k-NN model with decision trees. As a reminder, we had reached a precision of 92.8% with the KNN model.

As always, let's start by importing the high level features and labels.

````python
import os
import pandas as pd
import numpy as np

# Import train data
with np.load(os.path.join('processed_data', 'features_tr.npz'), allow_pickle=False) as data:
    features = dict(data.items())
X_tr = features['x']
y_tr = features['y']

# import validation data
with np.load(os.path.join('processed_data', 'features_va.npz'), allow_pickle=False) as data:
    features_va = dict(data.items())
X_va = features_va['x']
y_va = features_va['y']


#import raw test Images
with np.load(os.path.join('processed_data', 'features_te.npz'), allow_pickle=False) as data:
    features_te = dict(data.items())
X_te = features_te['x']
y_te = features_te['y']
````

## Accuracy with a depth of 3

Then, we create a simple decision tree with a depth of 3 that minimizes the Gini impurity measure at each split. What accuracy can we achieve? Is it better than with the KNN model?

````python
from sklearn.tree import DecisionTreeClassifier

# Decision tree
dt = DecisionTreeClassifier(max_depth=3, random_state=0, criterion ='gini')
dt.fit(X_tr, y_tr)

# Get score

# Get train/test scores
print('Train accuracy: {:.3f}'.format(
    dt.score(X_tr, y_tr)))

print('Test accuracy: {:.3f}'.format(
    dt.score(X_va, y_va)))
````

Let's try to plot our decision tree with ```graphviz```

## Plot the tree with graphviz

````python
from sklearn.tree import export_graphviz

# Export decision tree
tree_plot = export_graphviz(
    dt, # decision tree
    out_file=None,
    feature_names=None, class_names=['bike', 'car', 'motorcycle', 'other', 'truck', 'van'],
    filled=True, rounded=True, proportion=True   
)
````

````python
import graphviz

# Display decision tree
graphviz.Source(tree_plot)
````

We observe that the decision tree can quickly find the bikes thanks to the feature 1515 and 1405. But the decision tree is not able to create a classification for vans. 

## Reduce the number of dimensions with PCA first? 

With the Decision tree of depth 3 we obtain an accuracy of ~70%, which is less than the KNN. To improve our model, we could try to reduce the number of dimensions with the PCA and thus reduce the noise. Why this could work? 

Well, Entropy is used in calculating the Information Gain in Decision Trees, which is how Decision Trees are trained. Information Gain measures the difference in entropy if we were to create a decision point for a given dimension. For a binary feature, Information Gain would be the difference between the entropy of the original dataset, minus the entropy of dataset if we assume that feature is true, and the entropy of the dataset if we assume that feature is false. The original entropy is always larger, as the assumptions simplify the problem, decreasing the number of possibilities.

This makes PCA a natural fit to be applied before a Decision Tree is learned, as it explicitly transforms your dataset to highlight the directions that have the highest variance, which are often the directions that have the highest Information Gain while learning a Decision Tree. This makes the Decision Tree learn much faster, and achieve a high accuracy within the least number of decision layers.



````python
# Transform Data with PCA
from sklearn.decomposition import PCA

# Create PCA transformer
pca = PCA(n_components=175)

# Apply PCA to rescaled data
pca.fit(X_tr)
X_tr_pca = pca.transform(X_tr)
X_va_pca = pca.transform(X_va)
````

````python
dt.fit(X_tr_pca, y_tr)

# Get score
# Get train/test scores
print('Train accuracy: {:.3f}'.format(
    dt.score(X_tr_pca, y_tr)))

print('Test accuracy: {:.3f}'.format(
    dt.score(X_va_pca, y_va)))
````

We see that preprocessing data with PCA help with the accuracy of the model. We improve our accuracy from ~70% to 85%.
Let's see if different number of components can improve further the accuracy.

````python
# Generate a set of number of components
n_components = np.arange(1, 80, 1)

# Run several times the evaluation
gs_results = []
# Grid search
for n in n_components:
    # Create PCA transformer
    pca = PCA(n_components=n)
    pca.fit(X_tr)
    
    X_tr_pca = pca.transform(X_tr)
    X_va_pca = pca.transform(X_va)
    
    dt.fit(X_tr_pca, y_tr)

    # Save model and its performance on train/test sets
    gs_results.append({
        'number of components': n,
        'train_accuracy': dt.score(X_tr_pca, y_tr),
        'test_accuracy': dt.score(X_va_pca, y_va)
    })

# Convert results to DataFrame
gs_results = pd.DataFrame(gs_results)
gs_results.sort_values("test_accuracy", ascending = False).head()
````

Setting different number of components doesn't really improve our model. Moreover, the accuracy is still far from the results of KNN which was 92.8%.
