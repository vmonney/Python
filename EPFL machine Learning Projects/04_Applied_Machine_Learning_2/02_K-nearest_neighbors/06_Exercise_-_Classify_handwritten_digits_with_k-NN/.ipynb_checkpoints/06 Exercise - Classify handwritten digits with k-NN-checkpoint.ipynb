Classify handwritten digits with k-NN
---

Exercise - Load data set
--

> **Exercise**: Load the data from the `mnist-6k.npz` file into X/y arrays. The data type of the input matrix `X` should be `float64`. Split it into train (5,000 images) and test (1,000) sets using the `train_test_split()` function. Set the `random_state` to zero. 

```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.dummy import DummyClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.base import clone
```

```python
# Load data
with np.load('data/mnist-6k.npz', allow_pickle=False) as npzfile:
    # Load the arrays in a Python dictionary
    data = dict(npzfile.items())
    
X = np.array(data['data'], dtype=np.float64)
y = data['labels']

# Split into train/test sets
X_tr, X_te, y_tr, y_te = train_test_split(X, y, train_size = 5000, test_size=1000, random_state=0)
```

Exercise: Data exploration
---

> **Exercise**: What are the different categories? What is the proportion of images in each one?

The categories represent flattened pixel values of the image corresponding of the three RGB colors. So the images contains 784 white and black pixel or 28x28.

> **Exercise**: Plot some images with their labels from the X/y arrays.

```python
for i in range(1,10):
    img = X_tr[i].reshape(28, 28)
    labels = y_tr[i]
    # Plot the image
    plt.imshow(img, cmap=plt.cm.gray_r)
    plt.title(f'label:{labels}')
    plt.show()
```

Exercise: Set a baseline
---

> **Exercise**: Create a baseline model using the `DummyClassifier` object from Scikit-learn and evaluate its performance on the test set.

```python
# Compute distribution using Pandas
pd.Series(y_tr).value_counts(normalize = True)
```

```python
# Create the dummy classifier
dummy = DummyClassifier(strategy='most_frequent')

# Fit it
dummy.fit(X_tr, y_tr)

# Compute test accuracy
accuracy = dummy.score(X_te, y_te)
print('Baseline Accuracy: {:.3f}'.format(accuracy))
```

Exercise: Create k-NN classifier
---

> **Exercise**: Build a k-NN classifier using a `Pipeline` object (set k to 5). Try with and without standardization.

```python
# Create a k-NN pipeline with standardization
knn_pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('knn', KNeighborsClassifier(n_neighbors=5))
])

# Fit estimator
knn_pipe.fit(X_tr, y_tr)

# Evaluate on test set
accuracy = knn_pipe.score(X_te, y_te)
print('Accuracy: {:.3f}'.format(accuracy))
```

```python
# Create a k-NN pipeline without standardization
knn_pipe2 = clone(knn_pipe)
knn_pipe2.set_params(scaler=None)
knn_pipe2.fit(X_tr, y_tr)
print('Accuracy: {:.3f}'.format(knn_pipe2.score(X_te, y_te)))
```

Exercise: Tune k using grid search
---

> **Exercise**: Tune k using grid search. Plot the test accuracy.

**Hint**: You don't have to compute the train accuracy which is computationally expensive due to the size of the train set.

```python
# Variable to store the results
gs_results = []

# Grid search
for n in range(2,20):
    knn_pipe3 = Pipeline([
        ('scaler', StandardScaler()),
        ('knn', KNeighborsClassifier(n_neighbors=n))
    ])
    knn_pipe3.fit(X_tr, y_tr)
    accuracy = knn_pipe3.score(X_te, y_te)
    gs_results.append({
        'score': accuracy,
        'n_neighbors': n
    })

# Convert results to DataFrame
gs_results = pd.DataFrame(gs_results) 
gs_results.head()
```

```python
gs_results.sort_values('score',ascending = False).head(1)
```

```python
# Plot the validation curves
plt.plot(gs_results['n_neighbors'], gs_results['score'], label='test accuracy')
plt.xlabel('number of neighbors')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
