Exercise - Chigago gambling crimes
---

The city of Chicago disposes of 8 police patrols to fight gambling crimes. They want to assign each patrol to an area of Chicago where there is historically a lot of gambling crimes.

> **Exercise**: Load the gambling crimes data. Create an `X` array with the `Longitude` and the `Latitude` variables (drop the other ones). Create a map of the gambling crimes using these two variables. Find eight centers of gamblings crimes using *k*-means. Plot each cluster with a different color and mark their centers (the centroids).

```python
import numpy as np
import pandas as pd
import os
from PIL import Image

import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.cluster import KMeans
```

```python
# Import data
crimes_df = pd.read_csv(os.path.join('data', 'chicago-crimes.csv'))
crimes_df.head()
```

```python
# Extract longitude and latitude
X = crimes_df[['Longitude', 'Latitude']].values
# Plot the localisations
plt.scatter(X[:,0], X[:,1], s=6)
plt.title("Localisation of Chicago's gambling crimes");
```

```python
# Create k-means object
kmeans = KMeans(
    n_clusters=8,
    random_state=0, # Fix results
)

# Fit it to the data
kmeans.fit(X, y=None);

# Create figure
plt.figure(figsize=(10, 10)) # figure 10 by 10 inches

# Same scale for the x- and y-axis
plt.gca().set_aspect('equal', adjustable='box')

# Plot clusters
for cluster in np.arange(0,8,1):
    # Get points in this cluster
    idx = (kmeans.labels_ == cluster)
    
    # Plot points
    plt.scatter(
        X[idx, 0], # Third column: petal length
        X[idx, 1], # Fourth column: petal width
        label='cluster {}'.format(cluster),
        s=10
    )
    
    # Plot centroid
    centroid = kmeans.cluster_centers_[cluster]
    plt.plot(centroid[0], centroid[1], marker='*', color='black', markersize=12)

# Add legend and labels
plt.legend()
plt.title("Patrol areas to fight gambling crimes in Chicago")
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

> **Exercise**: Load the `pepper.jpg` image. Apply *k*-means to the image pixels to extract a palette of 3, 5 or 10 colors. Then re-draw the image using this new palette.

```python
# Load the image using PIL
pepper_img = Image.open('data/pepper.jpg')

# Load it into a Numpy array
img = np.array(pepper_img)

print('shape:', img.shape)
print('min:', img.min(), 'max:', img.max())
print('dtype:', img.dtype)
```

```python
# Plot the image
plt.imshow(img)
plt.show()
```

```python
# img is the (height, width, channels) array of pixels
X_2 = img.reshape(-1, 3) # -1 let numpy choose the correct value of n. 3 correspond to the RGB values
```

```python
def image_color(data, n_colors): 
    # Create a copy
    X_new = data.copy()
    
    # Create k-means object
    kmeans = KMeans(
        n_clusters=n_colors,
        random_state=0, # Fix results
    )

    # Fit it to the data
    kmeans.fit(X_new, y=None);
    
    # Select n most present colors and use it
    for color in np.arange(0,n_colors,1):
        # Get points in this cluster
        idx = (kmeans.labels_ == color)

        X_new[idx] = kmeans.cluster_centers_[color] # correspond to the RGB values of colors chosen

    new_img = X_new.reshape(img.shape)  

    # Plot the image
    plt.imshow(new_img/255) # Imshow expects pixels between 0 and 1 when dtype is float!
    plt.title("Number of colors: {}".format(n_colors))
    plt.show()
```

```python
image_color(X_2, 3)
image_color(X_2, 5)
image_color(X_2, 10)
image_color(X_2, 20)
plt.imshow(img)
plt.title("Original Image")
plt.show()
```
