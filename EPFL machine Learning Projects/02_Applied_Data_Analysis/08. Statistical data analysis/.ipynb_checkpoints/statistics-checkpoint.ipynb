```python
# Gaussian distribution
val2 = np.random.randn(N)
plt.hist(val2, 100, color='r', edgecolor='b')
plt.title("Histogram of random numbers drawn from the Gaussian distribution");
```

## The central limit theorem

```python
# define the number of random variables
N = 10000

# define the number of iterations
it = [1, 2, 4, 8]

f = plt.figure(figsize=(18, 10))
for i in range(4):
    # define an array of zeros which will holds the sums
    x = np.zeros(N)
    subplot = f.add_subplot(2,2, i+1)
    for j in range(N):
        for k in range(it[i]):
            x[j] = x[j]+ np.random.random()
    plt.hist(x, 100, density = True, color='r', edgecolor='b')
    plt.title('Number of iterations = '+str(it[i]))
    plt.setp(subplot.get_yticklabels(), visible =False)
    
plt.suptitle("Addition of uniform random variables converges to a Gaussian distribution", fontsize=20);
```

## Other measures: the z-score
The z-score is an essential value in statistics, and you might have heard of it already. Essentially it tells us how far away a specific data value is from the mean. For a data value x, the z-score is given by
z = x− ¯x / s where s is the standard deviation and ¯x is the mean. The z-score helps us normalize data and put it in context in terms of its mean. Let's compute the z -scores for our previous example.

```python
mean = np.mean(val3)
std = np.std(val3)
zscores=[]
for val in val3:
    z = (val-mean)/std
    zscores.append(z)
    
plt.bar(x_pos, zscores);
```

We say that the data set is now standardized because it has a mean of 0 and a standard deviation of 1. This can be a very useful property when we want to talk about the scales of different datasets. You will see later in the program that in machine learning it is extremely common to standardize data sets and there are prebuilt standardization packages that do all the work for us.

# 06. Point estimates and confidence intervals

```python
poisson1 = stats.poisson.rvs(mu=55, size=200000)
poisson2 = stats.poisson.rvs(mu=10, size=100000)
population = np.concatenate((poisson1, poisson2))
```

```python
population.mean()
```

```python
sample = np.random.choice(population, size = 1000)
sample_mean = sample.mean()
sample_mean
```

```python
data = (['red']*50000) + (['blue']* 30000) + (['green']*10000) +(['white']*10000)
```

```python
sample = random.sample(data, 100)

for color in set(sample): # the goal of set() is to select only unique value
    print( color + " proportion estimates:")
    print( sample.count(color)/100)
```

## Confidence intervals

```python
z = stats.norm.ppf(q = 0.975)
z
```

```python
pop_std = population.std()
pop_std
```

```python
margin_of_error = z * (pop_std / np.sqrt(1000))
margin_of_error
```

```python
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)
confidence_interval
```

```python
population.mean()
```

```python
# Lets try to calculate the 95% confidence interval for 10 samples
intervals =[]
sample_means = []

for sample in range(10):
    sample = np.random.choice(population, size=1000)
    sample_mean = sample.mean()
    sample_means.append(sample_mean)
    pop_std = population.std()
    margin_of_error = z * (pop_std/np.sqrt(1000))
    confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)
    intervals.append(confidence_interval)
```

```python
plt.figure(figsize=(12,9))

plt.errorbar(x=np.arange(0.1, 10, 1),
             y = sample_means,
             yerr = [(top-bot)/2 for top,bot in intervals],
             fmt='o')

plt.hlines(xmin =0, xmax=10, y=population.mean(), color="red", label="population mean")
plt.legend()
plt.title('95% confidence intervals for point estimates of the population mean')
plt.show()
```

#### What if we don't know the std of the population ? 

```python
sample_small = np.random.choice(population, size = 10)
sample_small_mean = sample_small.mean()
sample_small_mean
```

```python
# when we don't know the real population, we compute the t-critical value instead of the z-critical value
t = stats.t.ppf(q= 0.975, df=9)
t
```

```python
sigma = sample_small.std() / np.sqrt(10)
margin_of_error = t * sigma
confidence_interval = (sample_small_mean - margin_of_error, sample_small_mean + margin_of_error)
confidence_interval
```

```python
stats.t.ppf(q=0.975, df=9) - stats.norm.ppf(0.975)
```

```python
stats.t.ppf(q=0.975, df=999) - stats.norm.ppf(0.975)
```

```python
stats.t.interval(alpha = 0.95,               # Confidence level
                df = 9,                     # Degrees of freedom
                loc = sample_small_mean,    # Sample mean
                scale = sigma)              # Standard deviation estimate  
       
```

```python
# let's try different confidence intervall

for confidence in (.5, .75, .8, .90, .95):
    confidence_interval = stats.t.interval(alpha = confidence, df=9, loc = sample_small_mean, scale = sigma)
    interval = round(confidence_interval[1] - confidence_interval[0], 2)
    print(confidence, interval)
```

```python
p_value
```

In our case a p-value of 0.0018… is certainly less then our significance value of 0.05 therefore we can reject our hypothesis. If it was more that 0.05, we would have failed to reject the null hypothesis.

## Type I and II errors

- A __type I__ error occurs if we reject the null hypothesis when it is actually true (this is also callde a false positive)

- A __type II__ error occurs if we fail to reject the null hypothesis when it is actually false (this is know as false negative)

## Chi-square tests
The t-test we looked at before is great for comparing quantitative properties such as the mean. However when we are dealing with qualitative, or categorical data then we use a different test known as the chi-square test


```python
data = pd.DataFrame(['red']*50000 + ['blue']*30000 + ['green']* 10000 + ['white']*10000) 
sample = pd.DataFrame(['red']*600 + ['blue']*300 + ['green']*70 + ['white']*60) 
```

```python
#Test the null hypothesis that the sample has the same distributio as the population

data_count = pd.crosstab(index=data[0], columns="count")
data_count
```

```python
sample_count = pd.crosstab(index=sample[0], columns="count")
sample_count
```

```python
expected_count = data_count * len(sample) / len(data)
expected_count
```

```python
chi_square = (((sample_count - expected_count)**2) / expected_count).sum()
chi_square
```

```python
stats.chi2.ppf(q = 0.95, df = 3)
```

```python
p_value = 1 - stats.chi2.cdf(x=chi_square, df=3)
p_value
```

## Hypothesis test with random selected samples / t-test

```python
# Population
poisson1 = stats.poisson.rvs(mu=55, size=200000)
poisson2 = stats.poisson.rvs(mu=10, size=100000)
population = np.concatenate((poisson1, poisson2))
```

```python
test_sample = np.random.choice(population, size = 300)
test_sample
```

```python
t_statistic, p_value = stats.ttest_1samp(test_sample, popmean=population.mean())
t_statistic
```

```python
p_value
```

the p_value is 0.94 which is above the significance value of 0.05. We cannot therefore reject the null Hypothesis that the sample mean is the same as the population mean
## Hypothesis with random selected sample / Chi square

```python
data =pd.DataFrame(['red']*50000 + ['blue']*30000 + ['green']*10000 + ['white']*10000)
```

```python
sample_data = np.random.choice(data[0], size = 1030)
len(sample_data)
```

```python
data_count = pd.crosstab(index=data[0], columns = 'count')
data_count
```

```python
sample_count = pd.crosstab(index=sample_data, columns = 'count')
sample_count
```

```python
expected_count = data_count * len(sample_data) / len(data)
expected_count
```

```python
chi_square = (((sample_count - expected_count)**2) / expected_count).sum()
chi_square
```

```python
stats.chi2.ppf(q = 0.95, df = 3)
```

```python
p_value = 1 - stats.chi2.cdf(x=chi_square, df=3)
p_value
```

Here the p-value is less than the significance level, we can thus reject the null hypothesis, although it is true. (type I error).
