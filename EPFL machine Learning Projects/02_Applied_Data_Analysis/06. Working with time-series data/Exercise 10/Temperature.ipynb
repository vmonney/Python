# 10. Exercise: temperature data

```python
# Import libraries and load data
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

data = pd.read_csv('weatherHistory.csv', index_col='Date', parse_dates=True)
data = data.loc['2006':'2009', 'Temperature (C)']
data.tail()
```

```python
data.plot();
```

```python
# Create hourly and monthly mean temperature
hourly_mean_temp = data.groupby(data.index.time).mean()
monthly_mean_temp = data.groupby(data.index.month).mean()
monthly_mean_temp.index = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
```

```python
hourly_ticks = 4*60*60 * np.arange(6)
hourly_mean_temp.plot(xticks = hourly_ticks)
plt.ylabel('Mean Hourly Temperature');
```

```python
# setup plot
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14,5))

# plot monthly and hourly mean temperature
axes[0].set(title = 'Mean Hourly Temperature',
            xlabel = 'Hour',
            ylabel = 'Temperature (C)')
hourly_mean_temp.plot(ax=axes[0], xticks = hourly_ticks )

monthly_mean_temp.plot(ax=axes[1])
axes[1].set(title = 'Mean Monthly Temperature',
            xlabel = 'Month',
            ylabel = 'Temperature (C)')
fig.tight_layout()
plt.show()
```
