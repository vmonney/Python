Numpy polyfit function
---

```python
import pandas as pd

data_df = pd.read_csv('data-1.csv')
data_df.head()
```

```python
x = data_df.x.values
y = data_df.y.values

print('Type of x and y:', type(x), type(y))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

# Plot data points
plt.scatter(x, y)
plt.xlabel('x values')
plt.ylabel('y values')
plt.show()
```

```python
import numpy as np

coefs = np.polyfit(x, y, deg=1)
print('Coefficients:', coefs)
```

```python
# Extract a,b values
a, b = coefs

print('a:', a)
print('b:', b)
```

```python
# Generate a hundred values between 0 and 1
x_values = np.linspace(0, 1, num=100) # ~ [0, 0.01, 0.02, ... 0.98, 0.99, 1]

# Compute the associate y values
y_values = a*x_values + b

# Plot the line
plt.scatter(x, y) # Plot the data
plt.plot(x_values, y_values, c='C3', label='polyfit(deg=1)') # Prediction line
plt.legend()
plt.show()
```

```python
# Load second set of points
data_df = pd.read_csv('data-2.csv')
x2 = data_df.x.values
y2 = data_df.y.values
```

```python
# Your turn - plot data points!
plt.scatter(x2, y2)
plt.xlabel('x values')
plt.ylabel('y values')
plt.show()
```

```python
# Fit a line
coefs2 = np.polyfit(x2, y2, deg=1)
print('Coefficients:', coefs2)

# Generate a hundred values between 0 and 1
x_values2 = np.linspace(0, 1, num=100)

# Compute the associate y values
a, b = coefs2
y_values2 = a*x_values2 + b

# Your turn - plot the line!
# Plot the line
plt.scatter(x2, y2)
plt.plot(x_values2, y_values2, c = 'C3', label = 'polyfit(deg=1)')
plt.legend()
plt.show()
```

```python
# Fit a polynomial of degree 3
coefs2 = np.polyfit(x2, y2, deg=3)
print('Coefficients:', coefs2)
```

```python
a, b, c, d = coefs2
y_values2 = a*(x_values2**3) + b*(x_values2**2) + c*x_values2 + d
```

```python
# Predict y values
y_values2 = np.polyval(coefs2, x_values2)
```

```python
# Your turn - plot the curve with the points!
# Plot the line
plt.scatter(x2, y2)
plt.plot(x_values2, y_values2, c = 'C3', label = 'polyfit(deg=3)')
plt.legend()
plt.show()
```

```python
# Your turn - fit points from data-3.csv
# Load third set of points
data_df = pd.read_csv('data-3.csv')
x3 = data_df.x.values
y3 = data_df.y.values

# By increasing the function number of degrees
def fit(x, y, deg, xlabel, ylabel, legend):
    # Fit a polynomial
    coefs = np.polyfit(x, y, deg=deg)

    # Generate a hundred values between min(x) and max(x)
    x_values = np.linspace(min(x), max(x), num=100)

    # Predict y values
    y_values = np.polyval(coefs, x_values)

    # Plot curve
    plt.scatter(x, y)
    plt.plot(x_values, y_values, c='C3', label=legend)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend()
    plt.show()

fit(x3, y3, 2, 'x values', 'y values', 'polyfit(deg=2)')
```

## Applying the logarithm transformation

```python
fit(np.log(x3), y3, 1, 'log(x)', 'y values', 'polyfit(log(x), y, deg=1)')
```

### plot the model with the original scale

```python
# Fit a polynomial
coefs = np.polyfit(np.log(x3), y3, deg=1)

# Generate a hundred values between min(x) ans max(x)
x_values = np.linspace(x3.min(), x3.max(), num = 100)

# Predict y values
y_values = np.polyval(coefs, np.log(x_values))

# Plot curve
plt.scatter(x3, y3)
plt.plot(x_values, y_values, c='C3', label='polyfit(log(x), y, deg=1)')
plt.xlabel('x values')
plt.ylabel('y values')
plt.legend()
plt.show()
```
