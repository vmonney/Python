### Multicollinearity
Multicollinearity is when there is a strong correlation between independent variables. Linear regression or multilinear regression requires independent variables to have little or no similar features. Heatmap is an excellent way to identify whether there is multicollinearity or not. Multicollinearity is a situation where two or more predictors are highly linearly related. In general, an absolute correlation coefficient of >0.7 among two or more predictors indicates the presence of multicollinearity.

### Preprocessing function

Here, I create a function to process data, including a log transformation for numeric variables and the one-hot encoding technique for categorical variables. 

````python
# Preprocessing function
def preprocess(df):
    # Work on a copy
    df = df.copy()
    
    # Select only train data for modeling
    df_train = df[df['set'] == 'train']
    df_train = df_train.drop(['set'], axis=1)
    
    # Apply log-transform
    continuous = df_train.select_dtypes(include=[np.number]).columns
    df_train[continuous] = np.log1p(df_train[continuous])
    
    # One-hot encoding
    df_train = pd.get_dummies(df_train)  # With linear regression, you need to leave out one column.
                                                              # But if you estimate such models with regularization, for example ridge, 
                                                              # lasso or the elastic net, then you should not leave out any columns      
    return df_train

# Compute RMSE
def rmse(y, y_pred):
    return np.sqrt(MSE(y, y_pred))
````

## Model Fitting
### Simple model
Let's try to fit our first model. Here I need to select the two most important features. For that, I will use an object called ```SelectKBest``` which select features according to the F-value between feature for regression tasks. Next, we will use a linear regression model. To have more robust results, I will also use the cross-validation principle on 10 subsets.

#### Modeling Pipeline for linear regression

````python
# Define the linear regression function
def lin_reg(df, n_feat):

    # Transform with log for numeric and one-hot for categories
    preprocessed_df = preprocess(df)

    # Create X, y
    X = preprocessed_df.drop('SalePrice', axis=1)
    y = preprocessed_df.SalePrice.values

    # Select the two best features according to an f_regression
    selector = SelectKBest(f_regression, k=n_feat)
    selector.fit(X, y)

    # Get columns to keep and create new dataframe with those only
    cols = selector.get_support(indices=True)
    X_new = X.iloc[:,cols].values
    features.append(X.iloc[:,cols].columns)
    
    ## 10-folds Cross-Validation
    for run_idx in range(10):   
        # Split into train/test sets
        X_tr, X_te, y_tr, y_te = train_test_split(X_new, y, test_size=0.2, random_state=run_idx)

        # Standardize features
        scaler = StandardScaler()
        X_tr_scaled = scaler.fit_transform(X_tr)
        X_te_scaled = scaler.transform(X_te)
    
        # Linear regression
        lr = LinearRegression()
        lr.fit(X_tr_scaled, y_tr) # Fit to train data
        
        # Save model and its performance on train/test sets
        lr_results.append({
            'model': lr,
            'run_idx': run_idx,
            'train_rmse': rmse(y_tr, lr.predict(X_tr_scaled)),
            'train_mae': MAE(np.exp(y_tr) - 1 , np.exp(lr.predict(X_tr_scaled))-1),# Formula inverse of log1p == exp(x)-1
            'test_rmse': rmse(y_te, lr.predict(X_te_scaled)),
            'test_mae': MAE(np.exp(y_te) - 1 , np.exp(lr.predict(X_te_scaled))-1),# Formula inverse of log1p == exp(x)-1
            'baseline_mae': np.mean(np.abs((np.exp(y_te)-1) - (np.exp(np.median(y_tr))-1)))
        })
````

So after using the SelectKBest object, we will use only two variables: ```Gr Liv Area``` and ```TotalSF```

#### Simple Linear Regression

````python
lr_results = []
features = []
# computing the model
lin_reg(clean_data, 2)

# Convert results to DataFrame
lr_results = pd.DataFrame(lr_results)

# Compute train/test mean scores with std
mean_tr_rmse = lr_results.train_rmse.mean()
mean_te_rmse = lr_results.test_rmse.mean()
mean_tr_mae = lr_results.train_mae.mean()
mean_te_mae = lr_results.test_mae.mean()
baseline_mae = lr_results.baseline_mae.mean()

print('baseline MAE model: {:.1f}$'.format(baseline_mae))
print('mean train MAE linear regression: {:.1f}$'.format(mean_tr_mae))
print('mean train RMSE: {:.3f}'.format(mean_tr_rmse))
print('mean test  MAE linear regression: {:.1f}$'.format(mean_te_mae))
print('mean test  RMSE: {:.3f}'.format(mean_te_rmse))
print('features used:',features)
````

#### Modeling Pipeline for Ridge regression with several features

````python
# Define the linear regression function
def ridge_reg(df):

    # Transform with log for numeric and one-hot for categories
    preprocessed_df = preprocess(df)

    # Create X, y
    X = preprocessed_df.drop('SalePrice', axis=1)
    y = preprocessed_df.SalePrice.values
    
    # Grid search
    for alpha in np.linspace(1,50, num=20):

        # Number of feature selection    
        for n_feat in range(10,200):
            # Select the two best features according to an f_regression
            selector = SelectKBest(f_regression, k=n_feat)
            selector.fit(X, y)

            # Get columns to keep and create new dataframe with those only
            cols = selector.get_support(indices=True)
            X_new = X.iloc[:,cols].values
            features = X.iloc[:,cols].columns 

            # 10-folds Cross-Validation   
            for run_idx in range(10):   
                # Split into train/test sets
                X_tr, X_te, y_tr, y_te = train_test_split(X_new, y, test_size=0.2, random_state=run_idx)

                # Standardize features
                scaler = StandardScaler()
                X_tr_scaled = scaler.fit_transform(X_tr)
                X_te_scaled = scaler.transform(X_te)

                # Ridge regression
                ridge = Ridge(alpha= alpha)
                ridge.fit(X_tr_scaled, y_tr) # Fit to train data

                # Save model and its performance on train/test sets
                ridge_results.append({
                    'model': ridge,
                    'n_features': n_feat,
                    'alpha': alpha,
                    'run_idx': run_idx,
                    'train_rmse': rmse(y_tr, ridge.predict(X_tr_scaled)),
                    'train_mae': MAE(np.exp(y_tr) - 1 , np.exp(ridge.predict(X_tr_scaled))-1),# Formula inverse of log1p == exp(x)-1
                    'test_rmse': rmse(y_te, ridge.predict(X_te_scaled)),
                    'test_mae': MAE(np.exp(y_te) - 1 , np.exp(ridge.predict(X_te_scaled))-1),# Formula inverse of log1p == exp(x)-1
                })
````

#### Ridge Regression with several Features

````python
begin_time = dt.datetime.now()
ridge_results = []
features = []
# computing the model
ridge_reg(clean_data)

# Convert results to DataFrame
ridge_results = pd.DataFrame(ridge_results)

# Group results by number of features
gb_n_feat = ridge_results.groupby(['alpha','n_features'])

# Compute train/test mean scores with std for mae
mean_tr = gb_n_feat.train_mae.mean()
mean_te = gb_n_feat.test_mae.mean()
std_tr = gb_n_feat.train_mae.std()
std_te = gb_n_feat.test_mae.std()
n_feat = mean_tr.index.get_level_values(1).unique()
print("elapsed time:", dt.datetime.now() - begin_time)
````

#### Plot Ridge regression Results

````python
#Change background color
sns.set_style('white')

# Plot mean scores
for a in np.linspace(1,50, num=20):
    plt.plot(n_feat, mean_tr.xs(a, level='alpha', drop_level=False), label='train')
    plt.plot(n_feat, mean_te.xs(a, level='alpha', drop_level=False), label='test')

    # Quantify variance with ±std curves
    plt.fill_between(n_feat,
                     mean_tr.xs(a, level='alpha', drop_level=False) - std_tr.xs(a, level='alpha', drop_level=False), 
                     mean_tr.xs(a, level='alpha', drop_level=False) + std_tr.xs(a, level='alpha', drop_level=False),
                     alpha = 0.2)
    plt.fill_between(n_feat,
                     mean_te.xs(a, level='alpha', drop_level=False) - std_te.xs(a, level='alpha', drop_level=False), 
                     mean_te.xs(a, level='alpha', drop_level=False) + std_te.xs(a, level='alpha', drop_level=False),
                     alpha = 0.2)

# Add marker for best score
    best_n_feat = mean_te.xs(a, level='alpha', drop_level=False).idxmin()
    plt.scatter(best_n_feat[1], mean_te.xs(a, level='alpha', drop_level=False).min(), marker='x', c='red', zorder=10)

    best_result_ridge = gb_n_feat.get_group(best_n_feat)
    plt.title('Best n_feat: {} - alpha: {:.4f} - mae: {:,.0f}$'.format(
    best_n_feat[1], a, best_result_ridge.test_mae.mean()))

    plt.xlabel('Number of Features')
    plt.ylabel('MAE in $')
    plt.legend()
    plt.show()
````

### Complex Model

#### Modeling Pipeline for Lasso regression with "all" features

````python
# Define the linear regression function
def Lasso_reg(df):

    # Transform with log for numeric and one-hot for categories
    preprocessed_df = preprocess(df)

    # Create X, y
    X = preprocessed_df.drop('SalePrice', axis=1)
    y = preprocessed_df.SalePrice.values
    
    # Grid search
    for alpha in np.logspace(-4,-2, num=20):

        ## 10-folds CV
        for run_idx in range(10):  

            # Split into train/test sets
            X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.2, random_state=run_idx)

            # Standardize features
            scaler = StandardScaler()
            X_tr_scaled = scaler.fit_transform(X_tr.astype(np.float))
            X_te_scaled = scaler.transform(X_te.astype(np.float))

            # Ridge regression
            lasso = Lasso(alpha= alpha, max_iter=1e5)
            lasso.fit(X_tr_scaled, y_tr) # Fit to train data

            # Save model and its performance on train/test sets
            lasso_results.append({
                'model': Lasso,
                'alpha': alpha,
                'run_idx': run_idx,
                'train_rmse': rmse(y_tr, lasso.predict(X_tr_scaled)),
                'train_mae': MAE(np.exp(y_tr) - 1 , np.exp(lasso.predict(X_tr_scaled))-1),# Formula inverse of log1p == exp(x)-1
                'test_rmse': rmse(y_te, lasso.predict(X_te_scaled)),
                'test_mae': MAE(np.exp(y_te) - 1 , np.exp(lasso.predict(X_te_scaled))-1),# Formula inverse of log1p == exp(x)-1
            })
````

````python
lasso_results = []

# computing the model
Lasso_reg(clean_data)

# Convert results to DataFrame
lasso_results = pd.DataFrame(lasso_results)

# Group results by number of features
gb_alpha = lasso_results.groupby('alpha')

# Compute train/test mean scores with std for mae
mean_tr = gb_alpha.train_mae.mean()
mean_te = gb_alpha.test_mae.mean()
std_tr = gb_alpha.train_mae.std()
std_te = gb_alpha.test_mae.std()
alphas = mean_tr.index.values

# Plot mean scores
plt.plot(np.log10(alphas), mean_tr, label='train')
plt.plot(np.log10(alphas), mean_te, label='test')

# Quantify variance with ±std curves
plt.fill_between(np.log10(alphas), mean_tr-std_tr, mean_tr+std_tr, alpha=0.2)
plt.fill_between(np.log10(alphas), mean_te-std_te, mean_te+std_te, alpha=0.2)

# Add marker for best score
best_alpha = mean_te.idxmin()
best_result_lasso = gb_alpha.get_group(best_alpha).mean()
plt.scatter(np.log10(best_alpha), mean_te.min(), marker='x', c='red', zorder=10)
plt.title('Best alpha: {:.1e} - rmse: {:.4f} - mae: {:,.0f}$'.format(
    best_result_lasso.alpha, best_result_lasso.test_rmse, best_result_lasso.test_mae))

plt.xlabel('$log_{10}(alpha)$')
plt.ylabel('MAE')
plt.legend()
plt.show()
````
