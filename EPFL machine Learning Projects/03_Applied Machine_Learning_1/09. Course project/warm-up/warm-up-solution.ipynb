# Task 1

The goal of the first task is to fit the following equation to a set of 50 x/y data points: 

$$ y = a*log(x)+b $$

## Fit the curve and plot it

```python
# Import the libraries
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Load the data
df = pd.read_csv("task-1.csv")
df.head()
```

```python
# Create x, y 
x = df.x.values
y = df.y.values

# Plot the data
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))
    
ax[0].scatter(x, y)
ax[0].set_title("Normal data")

ax[1].scatter(np.log(x), y)
ax[1].set_title("log transformation on x axis");
```

We observe that by taking the log of the variable x, the relationship between x and y becomes linear. Let's fit the model accordingly.

```python
# fit the model 
coefs = np.polyfit(np.log(x), y, 1)

# Create x values
x_values = np.linspace(min(x), max(x), num=100)

# Make predictions for visualization
y_plot = np.polyval(coefs, np.log(x_values))

# Plot predictions
plt.scatter(x , y)
plt.plot(x_values, y_plot, c = 'C3', label = 'polyfit(log x),deg=1')
plt.legend();
```

The model seems to predict the targets correctly. But how well does it predict? Let's try to measure this with the $ R^2 $ formula.

## Compute the RSS and $ R^2 $ measures

```python
# Make the predictions for computing
y_pred = np.polyval(coefs, np.log(x))

# RSS
def RSS(y, y_pred):
    return np.sum(np.square(y - y_pred))

rss = RSS(y, y_pred)
rss_baseline = RSS(y, y.mean())

# R^2 coefficient                
R2 = 1 - (rss / rss_baseline)

# R^2
print("RSS value:{:.2f}".format(rss))
print("R^2 value:{:.2f}".format(R2))
```

## Is it a good $ R^2 $ score? 

We see that the $R^2$ value is 0.94, so the model performs way better than the baseline. In fact 94% of the variance in the target is explained by the variable x. 

However we have to be careful with the interpretation. First of all, we didn't separate the values into a train and test set. We fitted the model on the whole dataset, which in practice is not a good idea. Second, maybe our model is even better than that. But because it is trained on noisy data, the performance could be altered. Although it is a good metrics to measure performance in regression analysis, we should always complement that with other metrics our visual analysis to provide better interpretations.

# Task 2

The second task consists of fitting a model to a set of data points that contains outliers. We will try different approaches to handle outliers, and compare them with the MAE score on test set. Let's import the libraries and load the data.

```python
# Import the libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, HuberRegressor
from sklearn.metrics import mean_absolute_error as MAE
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set several visualization parameters for the notebook
sns.set_style('white')       # whitegrid
plt.rc('axes', titlesize=18)     # fontsize of the axes title
plt.rc('axes', labelsize=14)     # fontsize of the x and y labels
plt.rc('xtick', labelsize=13)    # fontsize of the tick labels
plt.rc('ytick', labelsize=13)    # fontsize of the tick labels
plt.rc('legend', fontsize=13)    # legend fontsize
plt.rc('font', size=13)          # controls default text sizes
```

```python
# Load the data
df = pd.read_csv("task-2.csv")
print(df.shape)
df.head()
```

In order to determine the outliers, we establish boxplots and scatter plots.

```python
# Create scatter and boxplots
features = ["x1", "x2", "x3"]

def distplot(data, x, y):
    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10, 6))
    fig.suptitle(x.name, fontsize=20, fontweight ='bold')
    
    sns.scatterplot(data=data, x=x, y=y, s=60, ax=ax1)
    ax1.set_title('Scatterplot')
    
    sns.boxplot(data=data, x=x, linewidth=2.5, ax= ax2)
    ax2.set_title('Boxplot')
    plt.show()
```

```python
for f in features:
    distplot(df, df[f], "y")
```

We notice that there are some outliers, because the values of the X columns are very compact. Let's try to detect outliers from the x axis using the z-score.

```python
# Compute z-scores and select outliers
idx = pd.DataFrame()
for f in features:
    z_scores = (df[f] - df[f].mean()) / df[f].std()
    idx[f] = np.abs(z_scores) > 2

# Plot outliers
for f in features:
    plt.scatter(df[~idx[f]][f], df[~idx[f]]["y"])
    plt.scatter(df[idx[f]][f], df[idx[f]]["y"], c='C3')
    plt.title(f)
    plt.show()
    
# Number of outliers
print("Total number of outliers:",idx.any(axis=1).sum())
```

There are a total of 49 outliers from the thousand values, I think we can remove them. Let's now test several regressions with and without outliers to see if it improves our model. Notice also the different scales from the features. We will need to standardize the data.

## Preprocessing Data

```python
# Create X, y
X = df.drop('y', axis=1).values
y = df.y.values

# Split into train/test sets
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.2, random_state=0)

# Standardize features
scaler = StandardScaler()
X_tr_rescaled = scaler.fit_transform(X_tr)
X_te_rescaled = scaler.transform(X_te)


################ Without outliers

df_wo = df[~idx.any(axis=1)]

# Create X, y
X_wo = df_wo.drop('y', axis=1).values
y_wo = df_wo.y.values

# Split into train/test sets
X_tr_wo, X_te_wo, y_tr_wo, y_te_wo = train_test_split(X_wo, y_wo, test_size=0.2, random_state=0)

# Standardize features
scaler = StandardScaler()
X_tr_rescaled_wo = scaler.fit_transform(X_tr_wo)
X_te_rescaled_wo = scaler.transform(X_te_wo)
```

## Fit Models and compare the test MAE score 

```python
# Linear regression
lr = LinearRegression()
lr.fit(X_tr_rescaled, y_tr) # Fit models to train data
y_pred = lr.predict(X_te_rescaled) # Predict the data
mae_lr = MAE(y_te, y_pred)

# Huber regression
lr_huber = HuberRegressor(epsilon = 1)
lr_huber.fit(X_tr_rescaled, y_tr) # Fit models to train data
y_pred_hub = lr_huber.predict(X_te_rescaled) # Predict the data
mae_lr_hub = MAE(y_te, y_pred_hub)

# Linear regression without outliers
lr2 = LinearRegression()
lr2.fit(X_tr_rescaled_wo, y_tr_wo) # Fit models to train data
y_pred_wo = lr2.predict(X_te_rescaled_wo) # Predict the data
mae_lr_wo = MAE(y_te_wo, y_pred_wo)

print("Linear regression with outliers, MAE score: {:,.2f}".format(mae_lr))
print("Huber regression with outliers, MAE score: {:,.2f}".format(mae_lr_hub))
print("Linear regression without outliers, MAE score: {:,.2f}".format(mae_lr_wo))
```

As you can see, The linear regression without outliers and huber regression models both perform better than the linear regression with outliers. The reason is that the Huber loss cost function is robust to outliers and smooth. It uses two formulas: below the treshold (epsilon) the loss function use a quadratic function like RSS, and above the treshold, it uses a linear loss function like the MAE. Be careful here because it could be misleading. Although we measure our performance with the MAE, the model is still trained to minimize the RSS for the linear regression example. That is why it does not perform well.

Again, we can see that removing outliers could be sometimes really benificial in comparison with other methods. But with several features, we often don't have the opportunity to see that, thus we use regularizations to make the models more robust to outliers.
